## Exercise 6: How to create great graphs with Matplotlib

1. Before you can start, you have to __import__ matplotlib.pyplot!

```python
import matplotlib.pyplot as plt
```

You consider an investment with the following cashflows:

- -5,000 USD today (Investment)
- 1,200 USD in 1 year 
- 1,500 USD in 2 years
- 2,000 USD in 3 years
- 2,500 USD in 4 years 
- 3,000 USD in 5 years
- 3,500 USD in 6 years

```python
# run the cell!
cf = [-5000, 1200, 1500, 2000, 2500, 3000, 2500]
cf
```

2. Create one __list with cumulative cashflows__ and one __list with timestamps__!

```python
cf_cum = []
cumul = 0
for i in cf:
    cumul += i
    cf_cum.append(cumul)
cf_cum
```

```python
x_axis = list(range(7))
x_axis
```

3. Visualize __cumulative cashflows__ in a simple __line plot__! (figure size: (12, 8)) (see result below)

```python
plt.figure(figsize = (12, 8))
plt.plot(x_axis, cf_cum)
plt.show()
```

4. Visualize __cumulative cashflows__ in a simple __scatter plot__! (figure size: (12, 8)) (see result below)

```python
plt.figure(figsize = (12, 8))
plt.scatter(x_axis, cf_cum)
```

5. __Customize__ the plot! (see final result below)

```python
#complete the code!

plt.figure(figsize = (12, 8))
plt.plot(x_axis, cf_cum, color = "red", marker = "D", markersize = 10, linestyle = "-.", label = "Cumulative Cashflows")
plt.grid()
plt.hlines(y = 0, xmin = x_axis[0], xmax = x_axis[-1], color = "blue", linestyles = "solid", label = "Break Even")
plt.title("Cumulative Cashflows Project B", fontsize = 15)
plt.xlabel("Time (in Years)", fontsize = 12)
plt.legend(loc = "best", fontsize = 15)
#plt.ylabel("CF", fontsize = 12)
plt.show()
```

```python
plt.figure(figsize = (12, 8))
plt.plot([-5000, -3800, -2300, -300, 2200, 5200, 7700], color = "red", linestyle = "-.", marker = "D", markersize = 10, label = "Cumulative Cashflows")
plt.grid()
plt.hlines(y = 0, xmin = x[0], xmax = x[-1], linestyles= "solid", color = "blue", label = "Break Even")
plt.title("Cumulative Cashflows Project B", fontsize = 15)
plt.xlabel("Time (in Years)", fontsize = 12)
plt.ylabel("CF (in USD)", fontsize = 12)
plt.yticks(range(-6000,8001,1000))
plt.xticks(x, labels = ["t0", "t1", "t2", "t3", "t4", "t5", "t6"])
plt.annotate("Break Even Point", xy = (3.1, 0),xytext = (2.5, 2000), arrowprops= {"color":"black"}, fontsize = 15)
plt.legend(loc = "best", fontsize = 15)
plt.show()
```
