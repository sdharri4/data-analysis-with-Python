## Getting IRR with a while loop

```python
cf = [-200,20,50,70,100,50]
```

```python
guess = 0.06
```

```python
f = 1 + guess
NPV = 0
for i in range(len(cf)):
    NPV += cf[i] / f**(i)
print(NPV)
```

```python
guess += 0.01
guess
```

### while loop

```python
guess = 0.06
f = 1 + guess
step = 0.0001
NPV = 0
for i in range(len(cf)):
    NPV += cf[i] / f**(i)
NPV
```

```python
while NPV > 0:
    guess += step
    f = 1 + guess
    NPV = 0
    for i in range(len(cf)):
        NPV += cf[i] / f**(i)
    print(NPV, guess)
print(NPV, guess)
```

```python
guess = 0.06
step = 0.0001
target_npv = 0
tolerance = 0.5

while True:
    f = 1 + guess
    NPV = 0
    for i in range(len(cf)):
        NPV += cf[i] / f**(i)
    #print(NPV, guess)
    diff = NPV - target_npv
    
    if abs(diff) > tolerance:
        if diff < 0:
            guess += step
        elif diff > 0:
            guess += step
    else:
        break
        
print(NPV, guess)
```
